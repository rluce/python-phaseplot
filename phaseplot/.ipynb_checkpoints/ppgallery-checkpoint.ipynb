```python
%matplotlib inline

import phaseplot
import matplotlib.pyplot as plt

# Plot a quadratic polynomial
phaseplot.phase_portrait(lambda(z): z*z - 1)
plt.show()
```

```python
# Oops, it looks like the plot range isn't large enough
phaseplot.phase_portrait(lambda(z): z*z - 1, box=[-2,2,-2,2])
plt.show()
```
